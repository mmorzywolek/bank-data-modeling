1. dane klientów banku:

# Modelowanie


```python
from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
from sklearn.metrics import accuracy_score, plot_roc_curve, confusion_matrix, precision_score, recall_score

from sklearn.tree import export_graphviz
from sklearn import tree
import graphviz
```

```python
#binaryzacja zmiennej celu
data['y_binarny'] = np.where(data['y'] == 'yes', 1, 0)
data = data.drop(['y'], axis=1)
```

```python
#kodowanie zmiennych objaśniających_kategorycznych
for col_ in col1:
    data = data.join(pd.get_dummies(data[col_], drop_first=True, prefix=col_)).drop(col_, axis=1)
```

```python
data.head()
```

```python
y = data['y_binarny']
X = data.drop('y_binarny', axis=1)
```

```python
#Podział na zbiór uczący i testowy
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

W tej części przystąpiono do zadania klasyfikacyjnego. Jako pierwszy model wybrano regresję logistyczną. Aby uniknąć nadmiernego dopasowania modelu do danych podczas optymalizacji hiperparametrów wykorzystano walidację krzyżową. Do oceny jakości klasyfikacji wykorzystano metrykę 'f1'. Celem graficznej reprezentacji dopasowania modelu do danych wygenerowano krzywą ROC oraz Confusion Matrix. W kolejnym kroku dokonano dodatkowej optymalizacji progu odcięcia.

## Model regresji logistycznej

```python
#Wybór modelu
model=LogisticRegression(random_state=0, solver='liblinear')

#Siatka parametrów
param_grid = {'C' : [0.001, 0.01, 0.1, 0.5, 1, 10]}
```

```python
#Dobór optymalnych parametrów (siatka parametrów;walidacja krzyżowa - 3 części)
clf=GridSearchCV(model, param_grid,cv = 3,scoring='f1')
```

```python
#Dopasowanie optymalnego modelu do zbioru uczącego
best_model = clf.fit(X_train,y_train)

#Parametry wybranego modelu
best_model.best_params_
```

```python
#Sprawdzenie modelu na zbiorze testowym
y_pred = np.where(best_model.predict_proba(X_test)[:,1] > 0.5, 1, 0)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
class_names=[0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="Blues" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Wybrane metryki dla regresji logistycznej")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("F1-score:",metrics.f1_score(y_test, y_pred))
```

```python
plot_roc_curve(best_model, X_test, y_test)
plt.title('Receiver Operating Characteristic')
```

Często próg odcięcia, ustalony na poziomie 0.5 nie jest optymalnym. Główną metryką był wybrany F1 -score. Dla kampanii marketingowej jest ważne nie tylko zadzwonić do osób, które z większym prawdopodobieństwem zapisują się na lokatę terminową, ale również nie tracić czasu na tych osób, które dokładnie odmówią. Według [[1]](https://arxiv.org/abs/1402.1892) optymalny próg odcięcia wynosi około 1/2 wyniku F1, który osiąga. Optymalny próg nigdy nie będzie większy niż 0.5. Wyniki dla różnych progów odcięcia są podane na rysunku również jak i inne metryki.

```python
def threshold_optima(model, table=False, plot=False):

    thresholds = np.linspace(0.1, 0.5, 41)
    results = pd.DataFrame(columns = ["threshold", "Accuracy", "Precision", "Recall", "F1-score", "ROC_AUC"])
    for threshold in thresholds:
        y_new_pred = (model.predict_proba(X_test)[:,1] >= threshold).astype(int)
        acc = metrics.accuracy_score(y_test, y_new_pred)
        prec = metrics.precision_score(y_test, y_new_pred)
        rec = metrics.recall_score(y_test, y_new_pred)
        f1 = metrics.f1_score(y_test, y_new_pred)
        roc_auc = metrics.roc_auc_score(y_test, y_new_pred)
        results = results.append(pd.Series([threshold, acc, prec, rec, f1, roc_auc], index=results.columns), ignore_index=True)
        
    if table != False:
        return results

    if plot != False:
        plt.figure(figsize=(12,8)) 
        plt.plot(results['threshold'], results['Accuracy'], marker='.', label='Accuracy', color='red')
        plt.plot(results['threshold'], results['Precision'], marker='.', label='Precision', color='blue')
        plt.plot(results['threshold'], results['Recall'], marker='.', label='Recall', color='green')
        plt.plot(results['threshold'], results['F1-score'], marker='o', label='F1-score', linewidth=3.5, color='magenta')
        plt.plot(results['threshold'], results['ROC_AUC'], marker='.', label='ROC_AUC', color='orange')
        plt.xlabel('Próg odcięcia')
        plt.ylabel('Wartość ')
        plt.legend()
        plt.show();  
```

```python
threshold_optima(best_model, plot=True)
```

```python
y_new_pred = (best_model.predict_proba(X_test)[:,1] >= 0.17).astype(int)
print("Nowe wybrane metryki dla regresji logistycznej")
print("Accuracy:",metrics.accuracy_score(y_test, y_new_pred))
print("Precision:",metrics.precision_score(y_test, y_new_pred))
print("Recall:",metrics.recall_score(y_test, y_new_pred))
print("F1-score:",metrics.f1_score(y_test, y_new_pred))
```

Po optymalizacji progu odcięcia przystąpiono dodatkowo do zadania klasyfikacyjnego z wykorzystaniem modelu drzewa decyzyjnego. Analogicznie do poprzedniego kroku dokonano walidacji krzyżowej celem dopasowania optymalnych hiperparametrów przy modelowaniu na zbiorze trenującym. Następnie dokonano oceny jakości klasyfikacji dla zbioru testowego za pomocą wskaźników liczbowych oraz wskaźników graficznych analogicznych jak w przy modelu regresji logistycznej tj. krzywa ROC oraz Confusion Matrix.

## Model drzewa decyzyjnego

```python
#Wybór modelu
model = DecisionTreeClassifier(random_state=0)

#Siatka parametrów
param_grid = {'max_depth' : [3, 4, 5],
              'min_samples_split' : [10, 20, 30],
              'min_samples_leaf' : [3, 6, 10]}
```

```python
#Dobór optymalnych parametrów (siatka parametrów;walidacja krzyżowa - 3 części)
clf=GridSearchCV(model, param_grid, cv = 3, scoring='f1')
```

```python
#Dopasowanie optymalnego modelu do zbioru uczącego
best_model=clf.fit(X_train, y_train)

#Parametry wybranego modelu
best_model.best_params_
```

```python
#Graficzna interpretacja drzewa decyzyjnego
classes = list(map(str,best_model.classes_.tolist()))
classes

dot_data = tree.export_graphviz(best_model.best_estimator_, out_file=None, 
                                feature_names=X.columns.values,  
                                class_names=classes,
                                filled=True)

graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
#Sprawdzenie modelu na zbiorze testowym
y_pred = best_model.predict(X_test)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
class_names=[0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="Blues" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Wybrane metryki dla drzewa decyzyjnego")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("F1-score:",metrics.f1_score(y_test, y_pred))
```
